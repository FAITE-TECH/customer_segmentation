```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
```

```python
df = df.dropna(subset=['CustomerID'])
df['CustomerID'] = df['CustomerID'].astype(int)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.to_csv("Online_Retail_Cleaned.csv", index = False)
```

```python
import datetime as dt

snapshot_date = df['InvoiceDate'].max() + dt.timedelta(days=1)
snapshot_date
```

```python
df['TotalPrice'] = df['Quantity']*df['UnitPrice']

rfm = df.groupby('CustomerID').agg({
    'InvoiceDate' : lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo' : 'nunique',
    'TotalPrice' : 'sum'
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']
```

```python
rfm.head()
```

```python
rfm.describe()
```

```python
rfm.to_csv("RFM_Table.csv")
```

```python
import numpy as np

for col in ['Recency', 'Frequency', 'Monetary']:
    upper_limit = np.percentile(rfm[col], 99)
    rfm[col] = np.where(rfm[col] > upper_limit, upper_limit, rfm[col])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)

import pandas as pd
rfm_scaled = pd.DataFrame(rfm_scaled, index = rfm.index, columns = rfm.columns)

rfm_scaled.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(rfm_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(K_range, inertia, 'bo-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

for k in range(2, 11):  # silhouette not valid for k=1
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(rfm_scaled)
    score = silhouette_score(rfm_scaled, kmeans.labels_)
    print(f"K={k}, Silhouette Score={score:.4f}")
```

```python
kmeans = KMeans(n_clusters=3, random_state =42, n_init=10)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)

print(rfm['Cluster'].value_counts())

rfm.groupby('Cluster').mean()
```

```python
cluster_summary = rfm.groupby('Cluster').mean().reset_index()
```

```python
rfm.to_csv("RFM_with_Clusters.csv")
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
pca_components = pca.fit_transform(rfm_scaled)

plt.figure(figsize=(8,6))
plt.scatter(pca_components[:, 0], pca_components[:, 1],
            c=rfm['Cluster'], cmap='viridis', alpha =0.6)

plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Customer Segments')
plt.colorbar(label='Cluster')
plt.show()
```

```python
cluster_summary = rfm.groupby('Cluster').mean()

cluster_summary.plot(kind='bar', figsize=(10,6))
plt.title('Cluster Profile by Average RFM Values')
plt.ylabel('Average Value')
plt.xticks(rotation=0)
plt.show()
```

```python
cluster_labels = {
    0: "At-Risk Customers",
    1: "Regular Customers",
    2: "VIP Customers"
}

rfm['Segment'] = rfm['Cluster'].map(cluster_labels)
rfm.to_csv("RFM_with_Segments.csv")
```

```python
segment_profile = rfm.groupby('Segment').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean', 'count']
}).round(2)

segment_profile.columns = ['Recency_Mean', 'Frequency_Mean', 'Monetary_Mean', 'Customer_Count']
segment_profile = segment_profile.reset_index()
segment_profile.to_csv("Segment_Profile.csv", index=False)
segment_profile
```
